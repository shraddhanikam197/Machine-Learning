```python
import pandas as pd
```

```python
df = pd.read_csv("Car_Data_v3.csv")
df.drop(["name", "torque"], axis = 1, inplace = True)
df.head()
```

```python
df.shape
```

```python
df.owner.unique()
```

```python
df = df[df.owner != "Fourth & Above Owner" ]
df.owner.unique()
```

```python
df.isnull().sum()
```

```python
df = df[~df.mileage.isnull()]
df
```

```python
df.isnull().sum()
```

```python
df.owner.unique()
```

```python
df.seller_type.unique()
```

```python
df["fuel"] = df.fuel.replace("LPG", "CNG")
df.fuel.unique()
```

```python
l = []
for i in df.year:
    a = 2023 - i
    l.append(a)
```

```python
df["year"] = l
df.head()
```

```python
l1 = []
for i in df.mileage:
    b = float(i.split(" ")[0])
    l1.append(b)
```

```python
df["mileage_kmpl"] = l1
```

```python
l2 = []
for i in df.engine:
    c = float(i.split(" ")[0])
    l2.append(c)
```

```python
df["engine_CC"] = l2
```

```python
df.head()
```

```python
df.drop(["engine", "mileage"], axis = 1, inplace = True)
```

```python
df.head()
```

```python
l3 = []
for i in df.max_power:
    d = i.split(" ")[0]
    l3.append(d)
```

```python
df["max_power_bhp"] = l3
```

```python
df.drop("max_power", axis = 1, inplace = True)
df.head()
```
